```python
def read_file(File_Name):
    MyFile = open(File_Name,"r")
    for line in MyFile.readlines():
        print(line)
    MyFile.close()
```

```python
def writte_on_file(File_Name,new_input):
        MyFile = open(File_Name,"a+")
        MyFile.write("\n"+new_input+"\n")
        MyFile.close()
```

```python
read_file("common_words.txt")
```
